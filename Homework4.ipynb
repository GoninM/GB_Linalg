```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)
```

```python
def kramer(A, b):
    det_A = np.linalg.det(A)
    det_Ai = []
    xi = []
    if not det_A == 0:
        n = len(A)
        m = len(A[0])
        for i in range(n):
            Ai = np.copy(A.T)
            Ai[i] = b
            Ai=Ai.T
            det_Ai = np.linalg.det(Ai)
            xi.append(det_Ai/det_A)
        return(xi)
    else:
        print('not ok')
    return(xi)
```

```python
A = np.array([[1,-2],[3,-4]])
b = np.array([1,7])
```

```python
kramer(A, b)
```
